```python
funnel = df.groupby('event_name')['user_id'].nunique()
funnel
```

```python
#Calculate Conversion Rates

views = funnel.get('view', 0)
add_to_cart = funnel.get('add_to_cart', 0)
purchase = funnel.get('purchase', 0)

view_to_cart = add_to_cart / views
cart_to_purchase = purchase / add_to_cart
overall_conversion = purchase / views

view_to_cart, cart_to_purchase, overall_conversion
```

```python
#Drop-off Analysis

drop_view_to_cart = 1 - view_to_cart
drop_cart_to_purchase = 1 - cart_to_purchase

drop_view_to_cart, drop_cart_to_purchase
```

```python
#Flags

df['view_flag'] = (df['event_name'] == 'view').astype(int)
df['cart_flag'] = (df['event_name'] == 'add_to_cart').astype(int)
df['purchase_flag'] = (df['event_name'] == 'purchase').astype(int)
```

```python
session_funnel = df.groupby('session_id').agg({
    'view_flag': 'max',
    'cart_flag': 'max',
    'purchase_flag': 'max'
}).reset_index()
```

```python
session_funnel.head()
```

```python
total_sessions = session_funnel.shape[0]

view_sessions = session_funnel['view_flag'].sum()
cart_sessions = session_funnel['cart_flag'].sum()
purchase_sessions = session_funnel['purchase_flag'].sum()

view_sessions, cart_sessions, purchase_sessions
```

```python
view_to_cart = cart_sessions / view_sessions
cart_to_purchase = purchase_sessions / cart_sessions
overall_conversion = purchase_sessions / view_sessions

view_to_cart, cart_to_purchase, overall_conversion
```

```python
session_funnel = session_funnel[
    (session_funnel['view_flag'] == 1) &
    (session_funnel['cart_flag'] <= session_funnel['view_flag']) &
    (session_funnel['purchase_flag'] <= session_funnel['cart_flag'])
]
```

```python
session_device = df.groupby('session_id')['device_type'].first().reset_index()
```

```python
session_funnel = session_funnel.merge(session_device, on='session_id', how='left')
```

```python
session_funnel.head()
```

```python
device_funnel = session_funnel.groupby('device_type').agg({
    'view_flag': 'sum',
    'cart_flag': 'sum',
    'purchase_flag': 'sum'
}).reset_index()
```

```python
#Conversion rates
device_funnel['view_to_cart'] = device_funnel['cart_flag'] / device_funnel['view_flag']
device_funnel['cart_to_purchase'] = device_funnel['purchase_flag'] / device_funnel['cart_flag']
device_funnel['overall_conversion'] = device_funnel['purchase_flag'] / device_funnel['view_flag']

device_funnel
```

```python
session_traffic = df.groupby('session_id')['traffic_source'].first().reset_index()

session_funnel = session_funnel.merge(session_traffic, on='session_id', how='left')
```

```python
#Aggregation
traffic_funnel = session_funnel.groupby('traffic_source').agg({
    'view_flag': 'sum',
    'cart_flag': 'sum',
    'purchase_flag': 'sum'
}).reset_index()
```

```python
#Conversion rates
traffic_funnel['view_to_cart'] = traffic_funnel['cart_flag'] / traffic_funnel['view_flag']
traffic_funnel['cart_to_purchase'] = traffic_funnel['purchase_flag'] / traffic_funnel['cart_flag']
traffic_funnel['overall_conversion'] = traffic_funnel['purchase_flag'] / traffic_funnel['view_flag']

traffic_funnel
```

```python
session_country = df.groupby('session_id')['country'].first().reset_index()

session_funnel = session_funnel.merge(session_country, on='session_id', how='left')
```

```python
country_funnel = session_funnel.groupby('country').agg({
    'view_flag': 'sum',
    'cart_flag': 'sum',
    'purchase_flag': 'sum'
}).reset_index()
```

```python
country_funnel['view_to_cart'] = country_funnel['cart_flag'] / country_funnel['view_flag']
country_funnel['cart_to_purchase'] = country_funnel['purchase_flag'] / country_funnel['cart_flag']
country_funnel['overall_conversion'] = country_funnel['purchase_flag'] / country_funnel['view_flag']

country_funnel
```

### 📱 Device-Based Funnel Insights

* **Tablet users have the highest view-to-cart conversion (28.4%)**, indicating strong product engagement at the browsing stage.
* **Mobile users have the highest overall conversion (1.76%)**, slightly outperforming desktop and tablet, suggesting relatively better end-to-end performance.
* **Desktop shows the lowest cart-to-purchase conversion (5.9%)**, indicating possible friction during checkout despite decent initial engagement.
* Overall, conversion rates across all devices are quite low (~1.6%–1.7%), suggesting systemic issues in the funnel rather than device-specific problems.


### 🌐 Traffic Source Insights

* **Ads traffic has the highest overall conversion rate (2.48%)**, making it the most effective channel in terms of driving purchases.
* **Organic traffic performs consistently well (2.01%)**, indicating high user intent and quality traffic.
* **Referral traffic performs the worst (0.75%)**, with extremely low cart-to-purchase conversion (2.87%), suggesting poor traffic quality or mismatched audience targeting.
* **Direct traffic shows moderate performance (1.55%)**, possibly representing returning users but with room for improvement in conversion.


### 🌍 Country-Based Funnel Insights

* **USA (3.36%) and UAE (3.39%) have the highest overall conversion rates**, significantly outperforming other countries, indicating strong purchasing behavior.
* **Germany shows strong top-of-funnel engagement (33.7% view-to-cart)**, but relatively lower purchase conversion, suggesting drop-offs in later stages.
* **Australia has zero purchases despite user activity**, indicating a critical issue such as payment barriers, pricing, or lack of trust.
* **India and Brazil show high engagement but extremely low purchase conversion**, suggesting potential issues like pricing sensitivity or payment limitations.
* Countries like **Japan, UK, and France show balanced funnel performance**, indicating stable and consistent user behavior.


### 🔥 Key Business Takeaways

* The biggest drop-off occurs in the **final stage (cart to purchase)** across most segments, indicating checkout or payment-related friction.
* High-performing regions (USA, UAE) suggest where scaling efforts (ads, marketing) could yield better ROI.
* Low-performing segments (referral traffic, certain countries) highlight areas where targeting, UX, or localization strategies need improvement.
